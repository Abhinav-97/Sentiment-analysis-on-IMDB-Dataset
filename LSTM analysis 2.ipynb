```python
import numpy as np
import keras
```

```python
from keras.datasets import imdb
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, MaxPooling1D, Convolution1D
from keras.layers.embeddings import Embedding
```

```python
from keras.preprocessing import sequence
```

```python
word_indices = imdb.get_word_index()
```

```python
word_indices
```

```python
top_words = 6000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=top_words)
```

```python
x_train = sequence.pad_sequences(X_train, maxlen=500)
x_test = sequence.pad_sequences(X_test, maxlen=500)
```

Creating a model with convolutional layer with an LSTM layer to learn spatial features

```python
model = Sequential()
model.add(Embedding(top_words,64, dropout=0.2))
model.add(Convolution1D(filters=64, kernel_size=3, padding="same", activation="relu"))
model.add(MaxPooling1D(pool_size=2))
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
print(model.summary())
```

```python
model.fit(x_train,y_train, epochs=5, batch_size=128)
```

```python
scores = model.evaluate(x_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

As we can see the accuracy is higher in this case and dimensionality is also less
