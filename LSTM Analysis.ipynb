Keras already has the iMdb dataset Built-in.The dataset 25000 training and testing reviews each. Also the data is formatted with each word replacing the rank of thier occurance 

```python
imdb.load_data()
```

get_word_index contains the dictionary of all the words with their occurance rank

```python
word_indices = imdb.get_word_index()
```

```python
word_indices
```

```python
INDEX_FROM=3   # word index offset
train,test = imdb.load_data(index_from=INDEX_FROM)
x_train, y_train =  train
x_test,y_test = test
```

```python
x_train.shape,x_test.shape
```

```python
word_to_id = {k:(v+INDEX_FROM) for k,v in word_indices.items()}
word_to_id["SPACE"] = 0
word_to_id["<START>"] = 1
word_to_id["END"] = 2
 
```

```python
id_to_word = {value:key for key,value in word_to_id.items()}
print(" ".join(id_to_word[1] for i in x_train[1]))
```

```python
top_words = 5000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=top_words)
```

# First model with One LSTM Layer without dropout

```python
model = Sequential()
model.add(Embedding(top_words,64))
model.add(LSTM(100))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.summary()
```

```python
x_train = sequence.pad_sequences(X_train, maxlen=500)
x_test = sequence.pad_sequences(X_test, maxlen=500)
```

```python
x_train.shape
```

```python
model.fit(x_train, y_train, epochs=3, batch_size=128)
```

```python
scores = model.evaluate(x_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

As we can see The validation accracy is 86.88 but the model is prone to overfitting hence we use droput layers

# Another model with 2 LSTM layers and also dropout layer

```python
model2 = Sequential()
model2.add(Embedding(top_words,64))
model2.add(LSTM(128, return_sequences=True, recurrent_dropout=0.1, dropout=0.2))
# model2.add(Dropout(0.2))
model2.add(LSTM(64, recurrent_dropout=0.1, dropout=0.2))
# model2.add(Dropout(0.2))
model2.add(Dense(1,activation="sigmoid"))
model2.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
model2.summary()
```

```python
model2.fit(x_train, y_train, epochs=5, batch_size=128)
```

Saving the model Using h5py

```python
model2.save("model2.h5")
model.save("model.h5")
```

```python
model1 = load_model("model.h5")
model2 = load_model("model2.h5")
```

```python
scores = model2.evaluate(x_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

The validation accuracy might be less but this model is less prone to overfitting and is a more generalized model
